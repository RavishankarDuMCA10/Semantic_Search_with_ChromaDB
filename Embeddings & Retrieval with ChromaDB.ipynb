## 2. <font color = red> ChromaDB </font>

### <font color = red> 2.1 Adding documents/embeddings to ChromaDB and working with Collections </font>

You can add embeddings or chunks of text directly to chroma using their collection methods.

* **`.create_collection()`**
* **`.list_collections()`**
* **`.delete_collections()`**









* **`collection.add()`**
* `collection.peek()`
* `collection.get()`

```python
# Pip install chromaDB

!pip install chromadb
```

```python
# Import ChromaDB and get the Chroma client

import chromadb
input_path_cromadb = '/content/drive/MyDrive/AI/IIITB GEN AI/Vector_Databases_&_Retrieval_Augmented_Generation/HelpMateAI_RAG_Files/Semantic_Search_using_CromaDB/Semantic_Search_with_Chromadb/'
chroma_client = chromadb.PersistentClient(path=input_path_cromadb)
```

```python
# Create a collection to store the embeddings. Collections in Chroma are where you can store your embeddings, documents, and any additional metadata.

collection = chroma_client.get_or_create_collection(name="Semantic_Search_with_Chromadb")
```

```python
# Add the documents, embeddings, and ids into the collection

collection.add(
    embeddings = section_chunk_embeddings['Embeddings'].to_list(),
    documents = section_chunk_embeddings['Chunk Text'].to_list(),
    ids = [str(i) for i in range(0, len(section_chunk_embeddings['Embeddings']))],
)
```

```python
# Peek at the initial elements of the collection

collection.peek()
```

```python
# Retrieve items from the collection

collection.get(
    ids = ['0','1']
)
```

```python
# Add only the documents and ids into the collection, and ChromaDB will create embeddings on its own using the default embedding model - 'all-MiniLM-L6-v2'

temp_collection = chroma_client.get_or_create_collection(name="Temporary")
documents = ['Generative AI is an exciting space!', 'Generative AI encompasses all modalities including LLMs', 'Can Generative AI take my job?']
collection.add(
    documents = documents,
    ids = [str(i) for i in range(0, len(documents))],
)
```

```python
# List all the collection created

chroma_client.list_collections()
```

```python
# Delete a collection

chroma_client.delete_collection("Temporary")
```

### <font color = red> 2.2 Querying on ChromaDB </font>

* **`collection.query()`**

```python
# Read the query from the user
query = input()

# Query the ChromaDB collection using `collection.query()
results = collection.query(
    query_texts=query,
    n_results=3,
    include = ['documents', 'distances']
)
```

### 1.3 <font color = red> Updating Collections </font>

* **`collection.update()`**
* **`collection.upsert()`**


```python
# Converting all text chunks to lowercase in section_chunk_embeddings DataFrame

section_chunk_embeddings['Chunk Text'] = section_chunk_embeddings['Chunk Text'].apply(lambda x: x.lower())
```

```python
# Adding a metadata column containing the title of the page and the section name

section_chunk_embeddings['Metadata'] = section_chunk_embeddings.apply(lambda x: {'Title': x['Title'], 'Section': x["Section Title"]}, axis=1)
```

```python
section_chunk_embeddings
```

```python
# Convert the metadata column to a list to feed it to ChromaDB

metadata_list = section_chunk_embeddings['Metadata'].tolist()
```

```python
# Show First data of `Metadata` from metadata_list
metadata_list[0]
```

```python
# Update the collection using `collection.upsert()`

collection.upsert(
    embeddings = section_chunk_embeddings['Embeddings'].to_list(),
    documents = section_chunk_embeddings['Chunk Text'].to_list(),
    metadatas = metadata_list,
    ids = [str(i) for i in range(0, len(section_chunk_embeddings['Embeddings']))],
)
```

```python
collection.peek()
```

```python
collection.get(
    ids = ['0','1']
)
```

```python
query = input()
```

```python
# Query on the collection again, this time with filters on the chunk texts and the metadata

results = collection.query(
    query_texts=query,
    n_results=3,
    where = {'Section': 'Critical response'},
    where_document = {"$contains": "positive"},
    include = ['documents', 'distances', 'metadatas']
)
```

```python
results
```

```python
data = {'Documents': results['documents'][0], 'Metadatas': results['metadatas'][0], 'Distances': results['distances'][0]}
results_df = pd.DataFrame.from_dict(data)
results_df
```

```python
# Query on the collection again, this time with filters on the chunk texts and the metadata without `where` and `where_document` parameters

results = collection.query(
    query_texts=query,
    n_results=3,
    # where = {'Section': 'Critical response'},
    # where_document = {"$contains": "positive"},
    include = ['documents', 'distances', 'metadatas']
)
```

```python
results
```

```python
data = {'Documents': results['documents'][0], 'Metadatas': results['metadatas'][0], 'Distances': results['distances'][0]}
results_df = pd.DataFrame.from_dict(data)
results_df
```

```python
# Query on the collection again, this time with filters on the chunk texts and the metadata
# And use logical operator like "$or", "$and" `where_document` and/or `where` parameters, currently using "$or" with [{"$contains": "positive"}, {"$contains": "review"}] to apply or condition in 'where_document'.

results = collection.query(
    query_texts=query,
    n_results=3,
    where = {'Section': 'Critical response'},
    where_document = {"$or": [{"$contains": "positive"}, {"$contains": "review"}]},
    include = ['documents', 'distances', 'metadatas']
)
```

```python
results
```

```python
data = {'Documents': results['documents'][0], 'Metadatas': results['metadatas'][0], 'Distances': results['distances'][0]}
results_df = pd.DataFrame.from_dict(data)
results_df
```
